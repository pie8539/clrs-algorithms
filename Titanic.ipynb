```python
# Standard imports

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import sklearn
```

```python
titanic = pd.read_csv('train.csv')
titanic.head()
```

```python
# Set with all features

X = titanic.drop(['Survived', 'PassengerId', 'Name', 'Ticket', 'Cabin'], axis=1)
X = pd.get_dummies(X, drop_first=True, dtype=int)
X.head()
```

```python
# Set with labels

Y = titanic['Survived']
Y.head()
```

```python
# Split to train and test data

from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=.2)

X_train.shape, X_test.shape
```

```python
# Selecting the algorithm

from sklearn.ensemble import RandomForestClassifier

clf = RandomForestClassifier()
```

```python
# fit the data

clf.fit(X_train, Y_train)
```

```python
# getting the test file and making it like X_train

test = pd.read_csv('test.csv')

test = test.drop(['PassengerId', 'Name', 'Ticket', 'Cabin'], axis=1)
test = pd.get_dummies(test, drop_first=True, dtype=int)
test.head()
```

```python
# making prediction and getting file

preds = clf.predict(test)

preds = pd.DataFrame(preds, columns=['Survived'])
preds['PassengerId'] = preds.index + 892

preds = preds[['PassengerId', 'Survived']]
preds = preds.set_index('PassengerId')
preds.head()
```
