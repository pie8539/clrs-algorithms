```python
# Standard imports

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import sklearn
```

```python
housing = pd.read_csv('train.csv')
housing.head()
```

```python
Y = housing['SalePrice']
Y.head()
```

```python
X = housing.drop(['SaleType', 'KitchenQual', 'GarageCars',	'GarageArea', 'Functional', 'SalePrice', '3SsnPorch', 'ScreenPorch', 'PoolArea', 'MiscVal', 'BsmtFinSF1',	'BsmtFinSF2',	'BsmtUnfSF',	'TotalBsmtSF', 'MSZoning', 'Utilities', 'Exterior1st',	'Exterior2nd', 'BsmtFullBath',	'BsmtHalfBath'], axis=1)
X = X.dropna(axis=1)
X = pd.get_dummies(X, drop_first=True, dtype=int)

X.head()

test = pd.read_csv('test.csv')

test = test.drop(['3SsnPorch', 'ScreenPorch', 'PoolArea', 'MiscVal', 'Electrical'], axis=1)
test = test.dropna(axis=1)
test = pd.get_dummies(test, drop_first=True, dtype=int)

test.head()

common_cols = list(set(X.columns) & set(test.columns))
X = X[common_cols]
test = test[common_cols]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=.2)

X_train.shape, X_test.shape
```

```python
from sklearn.ensemble import RandomForestRegressor

model = RandomForestRegressor()
```

```python
model.fit(X_train, Y_train)
```

```python
model.score(X_test, Y_test)
```

```python
preds = model.predict(test)

preds = pd.DataFrame(preds, columns=['SalePrice'])
preds['Id'] = preds.index + 1461

preds = preds[['Id', 'SalePrice']]
preds = preds.set_index('Id')
preds.head()
```

```python
preds.to_csv('preds.csv')
```
